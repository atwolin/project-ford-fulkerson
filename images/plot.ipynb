```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
N_values = [50, 100, 250, 500, 750, 1000, 1500, 2000]
v0_times = [109.825, 103.460, 119.181, 280.240, 489.802, 808.893, 1525.878, 2651.660]
v1_times = [10.892, 51.542, 176.492, 968.795, 2314.880, 4349.321, 8531.254, 15787.925]
v2_times = [118.912, 102.406, 128.363, 278.343, 492.010, 827.810, 1565.200, 2692.496]
```

```python
# Plot setup with broken axis
fig, (ax1, ax2) = plt.subplots(2, 1, sharex=True, gridspec_kw={'height_ratios': [1, 2]})
fig.subplots_adjust(hspace=0.1)

# Upper plot (break point above 1000)
ax1.set_xticks(N_values)
ax1.plot(N_values, v0_times, label="v0", marker='o')
ax1.plot(N_values, v1_times, label="v1", marker='o')
ax1.set_ylim(3500, 18000)
ax1.spines['bottom'].set_visible(False)
ax1.tick_params(labeltop=False, bottom=False)
ax1.set_ylabel("Time (ms)")
# Show value for each point on the upper plot
for i, txt in enumerate(v0_times):
    ax1.annotate(f'{int(txt)}', (N_values[i], v0_times[i]), textcoords="offset points", xytext=(0,10), ha='center')
for i, txt in enumerate(v1_times):
    ax1.annotate(f'{int(txt)}', (N_values[i], v1_times[i]), textcoords="offset points", xytext=(0,10), ha='center')


# Lower plot (up to 1000)
ax2.set_xticks(N_values)
ax2.plot(N_values, v0_times, label="v0", marker='o')
ax2.plot(N_values, v1_times, label="v1", marker='o')
ax2.set_ylim(0, 3000)
ax2.spines['top'].set_visible(False)
ax2.set_xlabel("N")
ax2.set_ylabel("Time (ms)")
# Show value for each point on the upper plot
for i, txt in enumerate(v0_times):
    if i == 0:
        ax2.annotate(f'{int(txt)}', (N_values[i], v0_times[i]), textcoords="offset points", xytext=(0,25), ha='center')
    elif i <=2:
        ax2.annotate(f'{int(txt)}', (N_values[i], v0_times[i]), textcoords="offset points", xytext=(0,35), ha='center')
    else:
        ax2.annotate(f'{int(txt)}', (N_values[i], v0_times[i]), textcoords="offset points", xytext=(0,10), ha='center')
for i, txt in enumerate(v1_times):
    if i == 0:
        ax2.annotate(f'{int(txt)}', (N_values[i], v1_times[i]), textcoords="offset points", xytext=(0,5), ha='center')
    elif i < 2:
        ax2.annotate(f'{int(txt)}', (N_values[i], v1_times[i]), textcoords="offset points", xytext=(0,10), ha='center')
    else:
        ax2.annotate(f'{int(txt)}', (N_values[i], v1_times[i]), textcoords="offset points", xytext=(0,10), ha='center')



# Break lines
d = .01  # size of break
# kwargs = dict(transform=ax1.transAxes, color='k', clip_on=False)
# ax1.plot((-d, +d), (-d, +d), **kwargs)
# ax1.plot((1 - d, 1 + d), (-d, +d), **kwargs)

# kwargs.update(transform=ax2.transAxes)
# ax2.plot((-d, +d), (1 - d, 1 + d), **kwargs)
# ax2.plot((1 - d, 1 + d), (1 - d, 1 + d), **kwargs)
kwargs = dict(marker=[(-1, -d), (1, d)], markersize=12,
              linestyle="none", color='k', mec='k', mew=1, clip_on=False)
ax1.plot([0, 1], [0, 0], transform=ax1.transAxes, **kwargs)
ax2.plot([0, 1], [1, 1], transform=ax2.transAxes, **kwargs)

# Legend
ax1.legend()

# Display plot
plt.show()
```

```python
# Plot setup with broken axis
fig, (ax1, ax2) = plt.subplots(2, 1, sharex=True, gridspec_kw={'height_ratios': [1, 2]})
fig.subplots_adjust(hspace=0.1)
ax1.set_xticks(N_values)

# Upper plot (break point above 700)
ax1.plot(N_values, v0_times, label="v0", marker='o')
ax1.plot(N_values, v2_times, label="v2", marker='o')
ax1.set_ylim(750, 3000)
ax1.spines['bottom'].set_visible(False)
ax1.tick_params(labeltop=False, bottom=False)
ax1.set_ylabel("Time (ms)")

# Lower plot (up to 700)
ax2.set_xticks(N_values)
ax2.plot(N_values, v0_times, label="v0", marker='o')
ax2.plot(N_values, v2_times, label="v2", marker='o')
ax2.set_ylim(80, 500)
ax2.spines['top'].set_visible(False)
ax2.set_xlabel("N")
ax2.set_ylabel("Time (ms)")


# Break lines
kwargs = dict(marker=[(-1, -d), (1, d)], markersize=12,
              linestyle="none", color='k', mec='k', mew=1, clip_on=False)
ax1.plot([0, 1], [0, 0], transform=ax1.transAxes, **kwargs)
ax2.plot([0, 1], [1, 1], transform=ax2.transAxes, **kwargs)

# Legend
ax1.legend()

# Display plot
plt.show()
```
